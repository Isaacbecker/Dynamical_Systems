# Dynamical Systems Terminology

For a function F, the orbit of $x_0$ under F is the infinite sequence $\{x_0,x_1,x_2,...\}$ where $x_n = F^n(x_0)$

A fixed point of a function $F$ is a point $x_o$ such that $F(x_0) = x_0$

A period point of a function F is a point $x_0$ such that $F^n(x_0)= x_0$ for some $n\geq=1$

A point $x_0$ is called eventually periodic under F if $x_0$ is not periodic but one (or more) of its iterates $F^n(x_0)$ is periodic

```python
# example: Doubling function

def doubling_function(x,n=1):
    """
    input: 
    x - linspace of values or int
    n - number of iterates
    
    output:
    return F^n(x)
    """
    # the doubling function
    f = lambda x: (2*x)%1
    
    #Iterations
    for i in range(n):
        x = f(x)
    return x
```

```python
n = 2
x = np.linspace(0,1,10000)
y = doubling_function(x,n)

plt.plot(x,y,',')
plt.plot(x,x,label = "Fixed Points")
plt.title("Doubling Function")
plt.xlabel("x")
plt.ylabel(f"$F^{n}(x)$")
plt.legend()

y = doubling_function(0.3,n)
```
